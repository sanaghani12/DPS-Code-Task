```python
from google.colab import drive
drive.mount('/content/drive')

import plotly.express as px
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/Data/220511_monatszahlenmonatszahlen2204_verkehrsunfaelle.csv")
```

```python
df = df[['MONATSZAHL', 'AUSPRAEGUNG','JAHR','MONAT','WERT']]
df = df.dropna()

fig = px.histogram(df, x="JAHR", y="WERT",
             color='MONATSZAHL', barmode='group',
             text_auto=True,
             height=400)
fig.update_layout(
            title={
            'text' : 'Number of Accidents per Category',
            'x':0.5
        })
fig.show()
```
